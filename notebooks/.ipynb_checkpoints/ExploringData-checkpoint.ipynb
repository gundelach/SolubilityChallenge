Begin by plotting the descriptors against S_o naively

```python
plt.scatter(df['MLOGP'],np.log(df['S0 (mM)']))
plt.scatter(df['ALOGP'],np.log(df['S0 (mM)']))
```

```python
reg = linear_model.LinearRegression()
X=np.array(df['MLOGP']).reshape(-1,1)
y=np.log(np.array(df['S0 (mM)'])).reshape(-1,1)
reg.fit(X,y)
reg.score(X,y)
```

```python
y_pred = reg.predict(X)
plt.scatter(df['MLOGP'],np.log(df['S0 (mM)']))
plt.plot(X,y_pred,color='blue',linewidth=3)
```

```python
desc_list=['MLOGP','ALOGP']
score_list=[]
for desc in desc_list:
    reg = linear_model.LinearRegression()
    X=np.array(df[desc]).reshape(-1,1)
    y=np.log(np.array(df['S0 (mM)'])).reshape(-1,1)
    reg.fit(X,y)
    score_list.append(reg.score(X,y))
    y_pred = reg.predict(X)
    plt.scatter(df[desc],np.log(df['S0 (mM)']))
    plt.plot(X,y_pred,linewidth=3)
    
    
plt.show()
print (score_list)
```

```python
def trim_axs(axs, N):
    """little helper to massage the axs list to have correct length..."""
    axs = axs.flat
    for ax in axs[N:]:
        ax.remove()
    return axs[:N]
```

```python
desc_list=['MW', 'Sv', 'Si', 'nROH', 'nOHp', 'nHDon', 'nHAcc', 'Hy', 'MLOGP', 'ALOGP']
score_list=[]
fig, axs = plt.subplots(5, 2, figsize=(15,15),sharey='row')
fig.subplots_adjust(hspace=0.4, wspace=0.0)

axs=trim_axs(axs,10)

for ax, desc in zip(axs, desc_list):
    reg = linear_model.LinearRegression()
    X=(np.array(df[desc]).reshape(-1,1))
    y=(np.array(df['S0 (mM)'])).reshape(-1,1)
    reg.fit(X,y)
    score = reg.score(X,y)
    score_list.append(score)
    y_pred = reg.predict(X)
    ax.scatter(X,(df['S0 (mM)']))
    ax.plot(X,y_pred,linewidth=3)
    title = 'Des: ' + desc + ', corr: ' + str(np.round(score,decimals=4))
    ax.set_title(title)
    
    
plt.show()
print (score_list)
```

Looking at the data above it is obvious, that the discrete descriptors of type functional groups count should not be processed using naive linear regression. Hence in a pure multi-linear linear regression model, only use continuous variables for now. The Hy descriptor appears to have no linear correlation with the data so no point including it, but should investigate further if has relationship of another form, or if an individual outlier is causing issues. Also exclude MLOGP for now as has corr lower than 0.1

### Need to clean up test set and perform unit conversion

```python
test_set = pd.read_csv('../data/raw/test_subset.csv')
```

```python
#df[df['model'].str.match('Mac')]
# find the strings which indicate lacking data and the index on which they occur
del_list = list(test_set[test_set['S (_g/L)'].str.match('TOO SOLUBLE')].index)
# get ridd of those indexes 
clean_test = test_set.drop(del_list)
# reset index count
clean_test = clean_test.reset_index(drop=True)
# changy type of float
clean_test['S (_g/L)']=clean_test['S (_g/L)'].astype('float64')
# creaty new column in units mM which is mmol/L
clean_test['S0 (mM)'] = clean_test['S (_g/L)'] * 1/clean_test['MW'] * 10**(3)
```

```python
X_test = np.array(clean_test.loc[:,cont_descriptors])
test_predict = reg.predict(X_test)
clean_test['predS']=test_predict
```

```python
clean_test['predS']-clean_test['S0 (mM)']
```

Need statistical measure of how bad this data is compared to the real values. 

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(clean_test['S0 (mM)'],clean_test['predS'])
```

```python
from sklearn.metrics import r2_score
r2_score(clean_test['S0 (mM)'],clean_test['predS'])
```

So this model is completely terrible
