```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import pprint as pp
```

```python
def akaike_inf_crit(k,y_test,y_pred):
    '''Calculates Akaikes information criterion given a number of params k, and a predicted and obsered data set'''
    residue = y_test - y_pred
    sse = (np.square(residue)).sum()
    return (2 * k + len(y_test) * math.log(sse))
```

```python
def model_pred_eval(y_pred,y_test,name='Model1'):
    '''Calculates statistical measures of merit for predictions vs observations, returns Pandas DataFrame'''
    stats_dict={}
    error = abs(y_pred - y_test)
    stats_dict['mean_abs_error']=[np.mean(error)]
    stats_dict['median_abs_error']=[np.median(error)]
    stats_dict['median_accuracy']= [100 - np.median(100 * (error/y_test))]
    stats_dict['R2']=[r2_score(y_test,y_pred)]
    stats_dict['RMSE']=[math.sqrt(mean_squared_error(y_test,y_pred))]
    stats_dict['min_error']=[error.min()]
    stats_dict['max_error']=[error.max()]
    return pd.DataFrame.from_dict(stats_dict,orient='index',columns=[name])
```

```python
# Credit: matplotlib documentation
# Source: https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/markevery_demo.html#sphx-glr-gallery-lines-bars-and-markers-markevery-demo-py
def trim_axs(axs, N):
    """little helper to massage the axs list to have correct length..."""
    axs = axs.flat
    for ax in axs[N:]:
        ax.remove()
    return axs[:N]
```

# Introduction
 
This assignment is about the 'Solubility Challenge', a challenge set by the cheminformatics community to predict the solubility of drug-like biomolecules based on a small, highly accurate data set of solubilities. Knowledge of the solubility of drug candidates is vital for effective drug development as insoluble drags are difficult to administer. Experimental determination is very challenging and hence there is a concerted effort do develop predictive models based for solubility based on structure information of biomolecules. 

To build a qualitative structure property relationship model for solubility, descriptors must be defined and calculated for the biomolecules in the training and testing dataset. For this assessment a training and a test data set were provided. In addition to the solubilities, 10 molecular descriptors, calculated using Dragon software, were provided. The training set contains 85 molecules and the test set contains 28 molecules with solubility data. The solubilities are derived form experiment yet since no source was provided, the quality/reliability of the data is not known. 

The training set is used to train statistical models. Their effectiveness is then tested on the test set. This ensures that the models are predictive, rather than simple reproducing the training set. None of the models in the training set may be in the test set. The training set is about $\frac{2}{3}$ of the data. This split allows a larger training set, at the expense of strength of validation the models via the test set. A small test set risks that 'non-conformist' molecules within may heavily skew the performance of predictive models. On the other hand, a small training set risks simplistic and non-representative models. A 50/50 split may provide a more balanced approach. 

A method of validating the choice of split would be to re-split the data using the same ratio and re-testing the models developed below. Ideally, the performance of the models should be independent of the splitting of the data. If this is the case, the split is valid. If not, the data set must be expanded, or a more even split ratio employed. No information is given on how the split was implemented. The data split must be unbiased, and hence a random split via a specified ration is recommended. Advance knowledge of the test set may lead to a biased sampling of the training set via previous knowledge of the molecules and their properties. For example, known outliers may be purposefully ignored and well behaved molecules included. 

The table below summarizes the 10 descriptors provided. ALOGP was used by Yalkowsky in early solubility models and will hence be an important descriptor. Yalkowsky also used melting point, which is strongly correlated to molecular weight for many molecules. Hence MW will also be an important factor. Sv and Si are also likely to be important as they are related to the inter-atomic forces the molecule would experience in solution. Intuitively the Hydrophilic factor should also be important, as it relates closely to the molecules interaction with water. A potential issue in the selection of descriptors is that they may be correlated. Especially MLOGP and ALOGP will be highly correlated and hence only one of the two measures should be employed. Alternatively methods like PCA and PLS can be used to form new derived descriptors which are orthogonal/independent.

# Data Pre-Processing
There we a number of bad entries in the original test set which did not have solubility values. These molecules were removed. Furthermore, the units of solubility did not match the training set and where hence converted. The processed test set was saved to a CSV file and is provided in the supporting material. 

```python
df = pd.read_csv('../data/raw/training_subset.csv') #load training set
clean_test = pd.read_csv('../data/processed/test_subset.csv') #load processed test set
target = 'S0 (mM)' # column name of solubility data
y_test = clean_test.loc[:,target].values # set solubility observations in test set
y = df.loc[:,target].values # set solubility observations in training set
```

# Multiple Linear Regression

## Investigating linear correlation of each descriptor to $\log(S)$
We begin by analyzing to what extent each predictor is linearly correlated to the solubility. This will allow identification of potential candidate predictors for multiple linear regression. Below, the pair wise linear regressions are plotted in order to visually inspect the data. The logarithm of solubility, $\log(S)$, is used as the dependent variable as this produces significantly stronger correlation with the descriptors MW, Sv, Si, MLOGP and ALOGP. The R2 value for each linear regression is shown in the title of each plot. The R2 value is the coefficient of determination and describes the proportion of the variance in the dependent variable which is predictable from the independent variables. A value close to 1 indicates a strong fit of the linear regression model to the data set. In the plots below, none of the descriptors exhibit a strong linear correlation to solubility. 

```python
desc_list=['MW', 'Sv', 'Si', 'nROH', 'nOHp', 'nHDon', 'nHAcc', 'Hy', 'MLOGP', 'ALOGP']
score_list=[]
fig, axs = plt.subplots(5, 2, figsize=(15,15),sharey='row') # configure plot
fig.subplots_adjust(hspace=0.4, wspace=0.0)
axs=trim_axs(axs,10)

for ax, desc in zip(axs, desc_list): # loop over descriptors and axis to generate plots
    reg = linear_model.LinearRegression() # initialize regression
    X=(np.array(df[desc]).reshape(-1,1)) # define X and y
    y=np.log((np.array(df['S0 (mM)'])).reshape(-1,1))
    reg.fit(X,y) # fit regression model
    score = reg.score(X,y) # calculate R2
    score_list.append(score)
    y_pred = reg.predict(X) # make prediction on training set X
    ax.scatter(X,(df['S0 (mM)'])) # plot scatter of X,y 
    ax.plot(X,y_pred,linewidth=3) # plot predicted linear model
    title = 'Des: ' + desc + ', corr: ' + str(np.round(score,decimals=4))
    ax.set_title(title) 
plt.show()
```

## Multiple linear regression using key descriptors 
Based on the plot above, we select all descriptors which have a linear correlation with R2>0.1 with $\log(S)$. MLOGP is discarded as is has lower R2 value than ALOGP and the two are highly correlated. This leaves the four descriptors MW, Sv, Si, and ALOGP. In general, the number of descriptors should be such that the number of observables is larger than $3^{\text{number of descriptors}}$. A choice of 4 descriptors satisfies this criterion with $3^4=81 < $ 85 observables. 
From these descriptors, we build a multiple linear least square regression model using least squares regression. The plots below illustrate the individual linear correlation between the chosen descriptors and the log solubility. 
 

```python
cont_descriptors=['MW','Sv','Si','ALOGP']
score_list=[]
fig, axs = plt.subplots(2, 2, figsize=(10,10),sharey='row')
fig.subplots_adjust(hspace=0.2, wspace=0.0)

axs=trim_axs(axs,10)

for ax, desc in zip(axs, cont_descriptors):
    reg = linear_model.LinearRegression()
    X=(np.array(df[desc]).reshape(-1,1))
    y=np.log((np.array(df['S0 (mM)'])).reshape(-1,1))
    reg.fit(X,y)
    score = reg.score(X,y)
    score_list.append(score)
    y_pred = reg.predict(X)
    ax.scatter(X,y)
    ax.plot(X,y_pred,linewidth=3)
    title = 'Des: ' + desc + ', corr: ' + str(np.round(score,decimals=4))
    ax.set_title(title)     
plt.show()
```

Below the model is fitted to the training data and then used to make prediction for the test set. The custom function `model_pred_eval` returns a dataFrame of statistical measures of the models performance on the test set. The function calculates:
    - R2 of the model on the test set
    - Root mean square error
    - mean absolute error
    - median absolute error
    - median accuracy as a percentage
    - mean square error 

```python
X_test=clean_test.loc[:,cont_descriptors].values
y_test=np.log(clean_test.loc[:,target].values)
X= df.loc[:,cont_descriptors].values
y= np.log(df.loc[:,target].values)
reg = linear_model.LinearRegression()
reg.fit(X,y)
score = reg.score(X,y)
y_pred = reg.predict(X_test)
print ('R2 on training set: ', score)
stats = model_pred_eval(y_pred,y_test,name='MLR_4')
ACI = akaike_inf_crit(4,y_test,y_pred)
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
print(stats)
print ('Akaike information criterion: ',ACI)
plt.scatter(y_pred,y_test), plt.show()
```

As demonstrated by the statistical measures in the output above, this model performs quite well. On the training set, the model has a strong R2 of 0.53. On the test set, the R2 however drops to 0.35. The scatter plot above shows the predicted versus true solubility of the test set. Visual inspection indicates a moderate linear trend, as reflected by the R2 value.

The mean absolute error is 1.82 log units yet the median absolute error is only 1.45. This indicates that most errors are small, with some larger outliers. This histogram of errors shown below confirms this distribution of errors. Furthermore, the errors appear to be normally distributed. This indicates that the descriptors are relatively independent and the regression is well behaved. The root mean square error (RMSE) is 2.54 log units and will be used as the main measure for comparing models.  

```python
error = (y_pred - y_test)
plt.hist(error,bins=6), plt.title("Absolute Error historgram"),plt.xlabel('Residues, log units')
plt.show()
```

## Multiple linear regression using all descriptors.
While above, only the 4 descriptors with the highest individual correlation with $\log(S)$ were used, including more descriptors may increase the quality of the multiple linear regression model. To test this, we build a multiple linear least square regression model using every descriptor. 

```python
desc_list=['MW', 'Sv', 'Si', 'nROH', 'nOHp', 'nHDon', \
           'nHAcc', 'Hy', 'MLOGP', 'ALOGP']
```

```python
X= df.loc[:,desc_list].values
y= np.log(df.loc[:,target].values)
X_test=clean_test.loc[:,desc_list].values
reg = linear_model.LinearRegression()
reg.fit(X,y)
score = reg.score(X,y)
y_pred = reg.predict(X_test)
print ('R2 on training set: ', score)
stats = model_pred_eval(y_pred,y_test,name='MLR_10')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
pp.pprint(stats)
ACI = akaike_inf_crit(10,y_test,y_pred)
print ('Akaike information criterion: ',ACI)
```

Due the increased number of parameters, the model should achieve a better fit on the training data. Indeed, the R2 value on the training set increases by 0.1 by including all descriptors, as does the R2 on the test set. The RMSE improved slightly to 2.28 log units. However, this does not necessarily indicate that the model is a better description of system. By including more descriptors the model may have become over-fitted to the training data. Furthermore, by incorporating more descriptors, the model complexity increases, and interpretability suffers. This can be described quantitatively by Akaike's information criterion (AIC), which uses the residual sum of squares to judge the goodness of a model but penalizes the score for increased number of parameters. Indeed, the ACI for the 4 parameter model above (AIC=153.4) is lower than the AIC for the 10 parameter model (AIC=159.5), indicating that using all parameters over-fits the test set and the 4 parameter model is more descriptive. 

# Principle component analysis on $\log(S)$
A major drawback of naive linear regression approaches as employed above is that the different descriptors may themselves be correlated. This can lead to poor performance of the linear regression, which assumes that all descriptors are independent. Principle component analysis (PCA) provides a means of negating this problem. PCA generates a set of new descriptors, called principle components, as linear combinations of the original descriptors, aiming to describe the maximum amount of variance in the data set with each principle component. The principle components are all orthogonal, i.e independent. Furthermore, only the leading principle components, describing most of the system variance, can be used, thus reducing the dimensionality of the problem. Below, a PCA analysis is set up and the first 8 principle components are used to build a principle component multiple linear least square regression model. Before constructing the PCs, the descriptors are scaled such that the mean of each descriptor is 0 and the standard deviation is 1. This allows a more balanced construction of the PCs than when the magnitudes of different descriptors vary greatly. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
desc_list=['MW', 'Sv', 'Si', 'nROH', 'nOHp', 'nHDon', 'nHAcc', 'Hy', 'MLOGP', 'ALOGP']
```

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
# Scale X using StandardScaler functinality 
normalize = StandardScaler()
normalize.fit(X)
X = normalize.transform(X)
X_test=normalize.transform(X_test)
```

```python
pca = PCA(n_components=8) # initialize PCA with 8 PCs
pca.fit(X) # fit PCA on scaled descriptors 
principalComponents =  pca.transform(X) # transfrom descriptors for training and test set
principalComponents_test = pca.transform(X_test)
print ('Ration of variance: ', pca.explained_variance_ratio_)
print ('Total variance covered: ', (np.array(pca.explained_variance_ratio_)*100).sum())
```

By using the first 8 PCs, 99.89% of the variance is described. The ration of variance above shows how the varianve is distributed between the PCs. 

```python
reg = linear_model.LinearRegression()
reg.fit(principalComponents,y)
score = reg.score(principalComponents,y)
y_pred = reg.predict(principalComponents_test)
print ('R2 on training set: ', score)
stats = model_pred_eval(y_pred,y_test,name='PCA8')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
pp.pprint(stats)
ACI = akaike_inf_crit(8,y_test,y_pred)
print ('Akaike information criterion: ',ACI)
plt.scatter(y_test,y_pred)
```

The PCR outperforms the simple multiple linear regresssion models both in predictive power and in ACI score. The R2 on the test set is 0.52 and almost equal to that on the training set. Furthermore, the RMSE is 2.17 log units, the best result so far. The PCR has only 8 parameters instead of the 10 of the over-parameterized multiple linear regression model, yet still outperforms it in every metric. This is also reflected by the lower ACI score. The scatter plot above of predicted verses actual solubilities in the test set shows a clear linear relationship. 

However, there is an additional factor to consider. The choice of number of PCs is somewhat arbitrary and the use of different numbers of PCs results in a worse models as shown below. Using 10 principle components decreases the quality of the model, despite the inclusion of more principle components. This could indicate, that the PCR with 8 PCs performs better due to random effects, rather than a systematic improvement over the model with 10 PCs. Using either 8 or 10 PCs risks over-fitting the model however using less PCs lead to significantly worse models as not enough variance is covered. 

This ties into the fact that we have access to the test set. This means that we can immediately check the validity of any model against the test set. Thus, the test set essentially becomes another training set because we use it to determine the best parameters (number of PCs) for the model. This would not be possible in a blind study in which the test set was not known in advance. As describe above, knowledge of the test sets allows fine tuning of the models that would otherwise not be possible and increase the risk of selecting a "lucky" model. On way to circumvent this problem would be to subdivide the training set into a smaller training and new test set. However, the training set provided here is much to small to employ this strategy. A build up or build down approach together with Akaike's information criterion would be a more systematic approach. ACI could also be applied on the training set in case the test set is not known.  

Lastly, a major drawback of PCA is the loss of interpretability of the model. It is not straightforward to decompose the PCR into contributions from the individual original descriptors. Thus, chemical insights may be low in the use of PCA and related methods. 

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
# Scale X using StandardScaler functinality 
normalize = StandardScaler()
normalize.fit(X)
X = normalize.transform(X)
X_test=normalize.transform(X_test)

pca = PCA(n_components=10) # initialize PCA with 8 PCs
pca.fit(X) # fit PCA on scaled descriptors 
principalComponents =  pca.transform(X) # transfrom descriptors for training and test set
principalComponents_test = pca.transform(X_test)
reg = linear_model.LinearRegression()
reg.fit(principalComponents,y)
score = reg.score(principalComponents,y)
y_pred = reg.predict(principalComponents_test)
stats = model_pred_eval(y_pred,y_test)
pp.pprint(stats)
ACI = akaike_inf_crit(10,y_test,y_pred)
print ('Akaike information criterion: ',ACI)
```

# Partial least squares regression on $\log(S)$
Similar to PCA, the Partial least squares (PLS) method decomposes the descriptors into new descriptors (latent variables) by projecting the descriptors and solubility to a new space. Unlike PCA, PLS aims to capture the variance in both the dependent and independent variables via the latent variables. As for the PCR above, multiple linear regression using PLS performs inconsistently for different numbers of components/descriptors. The strongest predictive model is achieved at 7 components, but it is not obvious why. Again, this may be a result of random effects and by choosing this model, I may be falling victim to publication bias. The statistical measures are all very similar to those obatined by the PCR with 8 components. However the ACI score is noticeably better as one parameter less is used. The same considerations regarding the availability of the test set apply to PLS as they did to PCA. 

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
# Scale X
normalize = StandardScaler()
normalize.fit(X)
X = normalize.transform(X)
X_test=normalize.transform(X_test)
#X = StandardScaler().fit_transform(X)
```

```python
pls = PLSRegression(n_components=4)
pls.fit(X,y)
y_pred=pls.predict(X_test)
score = pls.score(X,y)
print ('R2 on training set: ', score)
stats = model_pred_eval(y_pred,y_test,name='PLS7')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
pp.pprint(stats)
ACI = akaike_inf_crit(4,y_test,y_pred.flatten())
print ('Akaike information criterion: ',ACI)
plt.scatter(y_test,y_pred)
```

# Random Forest on $\log(S)$
So far, the PCR and PLS regression have produce the strongest models with comparable predictive power and ACI scores. In an attempt to improve upon these model, we try a completely different approach. Random Forest algorithms are a non-supervised ensemble learning method. During the training step, bootstrapping sampling techniques are used to generate "simulated ensembles" of observations from the training set. Each of these is used to train an individual decision tree. The predictive power of random forest method relies on averaging over the predictions of many decision trees. To avoid correlation between the individual decision trees, random subsets of descriptors are used in the individual decision trees. This method is popular as it essentially operates as a black box with few parameters and is noise resilient due to the averaging over many decision trees. The main drawback is the lack of interpetability of the resulting model. The individual trees make no meaningful predictions and it is difficult to extract physical meaning from the model. 

```python
from sklearn.ensemble import RandomForestRegressor
```

## Naive
First we implement a completely naive random forest regression on all unprocessed descriptors. The training set score is a measure of how well the algorithm has fit to the training data. A score close to 1 indicates a successful fitting. The RMSE value of the prediction of the naive random forest model is 2.80 which is worse than the PCR/PLS methods above as well as the simple multiple linear regression. The R2 score is notably low. This may be because the random forest method is not a linear model and thus the quality of its fit may not be best measured using a linear measure like R2. Furthermore, random forest method is non-parametric as it makes no assumptions about underlying probability distributions. This implies that non-parametric statistical tests would be required to properly probe the generated models. The RMSE is valid for non-linear models and will thus be used as the mean measure of quality for the random forest and neural net methods. 

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
rf.fit(X,y)
y_pred = rf.predict(X_test)
print ('Score on training set', rf.score(X,y) )
stats = model_pred_eval(y_pred,y_test,name='Forest_Naive')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')

pp.pprint(stats)
```

## PCA and scaling random forest
To improve the results of the random forest method, the descriptors are scaled and then a principle component analysis is applied to ensure othogonaliyt/independence of the input descriptors, which is especially important for random forest algorithms. Applying these methods does notably increase the quality of the random forest model yet the RMSE of 2.55 is still larger than that of PCR and PLS models. Like all unsupervised machine learning methods, the random forest algorithm is very data hungry and would likely benefit more from a larger training set than the multiple linear regression methods above.

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
# Scale X
normalize = StandardScaler()
normalize.fit(X)
X = normalize.transform(X)
X_test=normalize.transform(X_test)
```

```python
pca = PCA(n_components=10)
pca.fit(X)
principalComponents =  pca.transform(X)
principalComponents_test = pca.transform(X_test)
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = 41)
rf.fit(principalComponents,y)
y_pred = rf.predict(principalComponents_test)
print ('Score on training set', rf.score(principalComponents,y) )
stats = model_pred_eval(y_pred,y_test,name='Forest_PCA10')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
pp.pprint(stats)
plt.scatter(y_pred,y_test)
```

# Neural Net with PCA
The last method attempted is the use of neural network based regression. This is a deep learning method which allows extremely complex, non-liner, functional forms to which to fit the data. The functional form is determined by the architecture of the neural net. The algorithm works by assigning many parameters to the complex functional form by minimizing against the training set. Due to the complexity of the minimization task, only local minima are found and hence the performance of the neural net will vary for each re-training. The number of parameters of the neural net must not exceed the number of observations, else the model will be over-fitted. The training set has 85 observations. Hence a neural net architecture with 8 PC inputs and three hidden layers of 4 hidden nodes each was chosen. The total number of parameters is $8\times 4 + 4 \times 4 + 4\times 4 + 4 \times 1 = 68$ parameters. The maximum iterations were set to 50000 to ensure convergence on the training set. 


```python
from sklearn.neural_network import MLPRegressor
```

```python
X = df.loc[:,desc_list].values
X_test=clean_test.loc[:,desc_list].values
# Scale X
normalize = StandardScaler()
normalize.fit(X)
X = normalize.transform(X)
X_test=normalize.transform(X_test)
```

```python
pca = PCA(n_components=8)
pca.fit(X)
principalComponents =  pca.transform(X)
principalComponents_test = pca.transform(X_test)
```

```python
mlp = MLPRegressor(max_iter=50000,hidden_layer_sizes=(4,4,4))
mlp.fit(principalComponents,y)
y_pred = mlp.predict(principalComponents_test)
stats = model_pred_eval(y_pred,y_test,name='NeuralNet')
stats_all = pd.DataFrame.join(stats,stats_all,lsuffix='_')
pp.pprint(stats)
plt.scatter(y_pred,y_test)
```

Them main problem with the neural network models is that they perform extremely inconsistently. On many runs, the neural net predicts massive outliers with absolute errors larger than 20 log units. On other runs, the neural net performs well and and RMSE score of about 3.0 log units is achieved. The main reason for this is likely the tiny size of the training and test sets. Neural nets excel when fed huge amounts of data. 

In an attempt to overcome this, I tried to use the training score generated after training of the neural net to predict which networks would perform best. Using only 10 PCs a smaller neural net with 4 hidden layers of 3,3,3 hidden nodes, 200 neural nets are trained on the training set and their performance on the test set saved. The warnings indicate that not all optimizations on the training set converged. 

```python
pca = PCA(n_components=8)
pca.fit(X)
principalComponents =  pca.transform(X)
principalComponents_test = pca.transform(X_test)
```

```python
len(y_test)
```

```python
list_of_predictions=[]
y_pred_total=np.array([0.0]*28)
for i in range(1,200):
    mlp = MLPRegressor(max_iter=8000,hidden_layer_sizes=(4,4,4))
    mlp.fit(principalComponents,y)
    y_pred = mlp.predict(principalComponents_test)
    y_pred_total += y_pred
    stats = model_pred_eval(y_pred,y_test)
    stats['score']=mlp.score(principalComponents,y)
    list_of_predictions.append(stats)
```

```python
stats = model_pred_eval(y_pred_total/199,y_test)
stats
```

```python
plt.scatter(y_pred_total/199,y_test)
```

As evident by the output blow, the highest training score does not necessarily result in the best fit to the test set. However, by training many neural networks and choosing to use only those with the highest training scores, the overall consistency of predictions on the test set can be improved somewhat. One potential approach could be to use the average predictions of the neural nets with the highest training scores. Lastly, the architecture of the neural net is likely to have a significant impact on the performance of the models generated. A deeper understanding of neural networks would however be required to fine tune the architecture of the net for this specific task.
